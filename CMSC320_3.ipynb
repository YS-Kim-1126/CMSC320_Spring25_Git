## Q2) Cleaning the Dataset (25 PTS Total)

**TASK 2.1 (10 Points):** Lets begin by fixing the Birthday column. You may notice that the dates are in inconsistent format!

- Pandas internally represent dates in 'datetime64' format which is a form of ISO 8601. Basically, dates are represented as yyyy-mm-dd.


- Fix all the dates in the Birthday Column.
- Display the "Profiles" dataframe after you are done.
- Note that the only **2 formats** the dates are in is either **yyyy/dd/mm** or **mm/dd/yyyy**;
- You want to represent them as **yyyy-mm-dd**

- **EXAMPLE**: In the ‚ÄúProfiles‚Äù dataset, dates like *2000/10/12* and *10/12/2000* follow different formats‚Äîone as *December 10th, 2000*, and the other as *October 12th, 2000*. We aim to standardize them to *2000-12-10* and *2000-10-12*, respectively.

**HINT:** some datetime functions might make things easier.
- https://www.geeksforgeeks.org/python-pandas-to_datetime/
- https://www.programiz.com/python-programming/datetime/strftime


```python
def standardize_date(date):
  date = str(date)
  try:
    date = datetime.strptime(date, "%Y/%d/%m")
    return date.strftime("%Y-%m-%d")
  except ValueError:
    pass

  try:
      date_obj = datetime.strptime(date, "%m/%d/%Y")
      return date_obj.strftime("%Y-%m-%d")
  except ValueError:
      pass
# Apply the function

profiles_df['Birthday'] = profiles_df['Birthday'].apply(standardize_date)
profiles_df
```

```python
# Find Alan Turing's message
alan_turing_row = interrogation_df[interrogation_df['Name'] == 'Alan Turing']
encoded_message = alan_turing_row['Statement'].values[0].strip('"')
print("\n{'Encoded Message':", encoded_message, "}")

# Decode the message properly with word spacing
number_groups = encoded_message.split('  ')  # Split on double space for words
decoded_message = []

for group in number_groups:
    word = []
    numbers = group.split()  # Split single space for letters
    for num in numbers:
        # Convert number to letter (A=1, B=2, etc.)
        word.append(chr(int(num) + 64))
    decoded_message.append(''.join(word))

decoded_message = ' '.join(decoded_message)
print("\n{'Decoded Message':", decoded_message, "}")
```

**TASK 4.3 (10 Points):**

**Restore** the interrogation statements back to their original state after **following the decoded message from before**.
- After restoration, make sure your modifications are updated in the **interrogation_df** itself.
- **Display** the **interrogation_df** when you are done.

```python
def clean_statement(statement):
    # Remove symbols after each vowel (a, e, i, o, u)
    vowels = {'a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U'}
    cleaned = []
    i = 0
    while i < len(statement):
        cleaned.append(statement[i])
        if statement[i] in vowels:
            # Skip next symbol if it's one of the special characters
            if i+1 < len(statement) and statement[i+1] in {'%', '$', '^','@',}:
                i += 1
        i += 1
    return ''.join(cleaned)

# Apply cleaning to all statements
interrogation_df['Statement'] = interrogation_df['Statement'].apply(lambda x: clean_statement(x.strip('"')))
display(interrogation_df)
```
